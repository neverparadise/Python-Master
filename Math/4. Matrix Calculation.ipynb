```python
class Matrix:
    def __init__(self, matrix_list, dtype):
        self.matrix = matrix_list
        self.dtype = dtype
        for _1, row in enumerate(self.matrix):
            for _2, col in enumerate(row):
                if type(col) != self.dtype:
                    print(f"The item {_} isn't same data type...")
                    raise ValueError
    
    def check_dimension(self, other):
        if (len(self.matrix[0]) != len(other.matrix[0]) and 
            len(self.matrix) != len(other.matrix)):
            print(f"The dimension of two matrixs are not same")
            return False
        else:
            return True
    
    def __add__(self, other):
        matrix = []
        if self.check_dimension(other):
            for row1, row2 in zip(self.matrix, other.matrix):
                row = []
                for col1, col2 in zip(row1, row2):
                    row.append(col1 + col2)
                matrix.append(row)
        else:
            raise ValueError
        
        new_matrix = Matrix(matrix, self.dtype)
        return new_matrix
    
    def __sub__(self, other):
        matrix = []
        if self.check_dimension(other):
            for row1, row2 in zip(self.matrix, other.matrix):
                row = []
                for col1, col2 in zip(row1, row2):
                    row.append(col1 - col2)
                matrix.append(row)
        else:
            raise ValueError
        
        new_matrix = Matrix(matrix, self.dtype)
        return new_matrix
    
    def __mul__(self, other):
        matrix = []
        for i in range(len(self.matrix)):
            row = []
            for j in range(len(other.matrix[0])):
                product = 0
                for k in range(len(other.matrix)):
                    product += self.matrix[i][k] * other.matrix[k][j]
                row.append(product)
            matrix.append(row)
        new_matrix = Matrix(matrix, self.dtype)
        return new_matrix
    
    def multiplication(self, other):
        matrix = []
        for this_row in self.matrix:
            new_row = []
            for other_col in other.matrix:
                product = 0
                for k in range(len(other.matrix)):
                    product += self.matrix[i][k] * other.matrix[k][j]
                row.append(product)
            matrix.append(row)
        new_matrix = Matrix(matrix, self.dtype)
        return new_matrix
    
    def __str__(self):
        msg = ""
        for row in self.matrix:
            msg += f"{row} \n"
        return msg
```
